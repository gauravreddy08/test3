```python
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## More details in the suggested reading for the discussion

```python
XY_MEAN = np.array([0, 0])
XY_COV = np.array([[2.2, 1],
                   [1, 1]])
```

```python
P_XY = multivariate_normal(mean=XY_MEAN, cov=XY_COV)
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

fig, axs = plt.subplots(2, 1)
fig.set_size_inches(6, 12)
axs[0].contourf(x, y, P_XY.pdf(pos))
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Dist P_XY')

xy_samp = P_XY.rvs(10000, random_state=42)
x_samp, y_samp = xy_samp[:, 0], xy_samp[:, 1]
z_samp = x_samp - 2 * y_samp
axs[1].hist(z_samp, bins=50, density=True)
axs[1].set_xlabel('Z')
axs[1].set_title('Dist P_Z')
```
